```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns

# import statsmodels.formula.api as smf

fb = pd.read_csv('DATA/dataset_Facebook.csv', delimiter=";")

```

```python
sharena = fb.share.isna()
fb[sharena]
```

```python
fb = facebook[facebook.share.notna()]
fb = fb[facebook.Paid.notna()]
fb.info()
```

```python
fb.reset_index(inplace=True)
fb
```

```python
grouped = fb.groupby('Post Weekday')
grouped = grouped.aggregate(np.mean).reset_index()
grouped
```

```python
sns.barplot(x = 'Post Weekday', y ='Lifetime Post Consumers', data = fb, hue = 'Paid', ci = None)
```

```python
sns.set(rc={'figure.figsize': (14, 8)})
sns.set_style('ticks')

ax = sns.boxplot(x = 'Post Weekday', y = 'Lifetime Post Consumers', data = fb, hue = 'Paid')
```
